```python
import numpy as np
#1-a

matriz = np.array ([
    [2,3],
    [2,3.1]
])

# Calculo la norma infinito
norma = np.linalg.norm(matriz, ord=np.inf)

# Calculo la inversa
inversa = np.linalg.inv(matriz)

# Calculo la nomra infinit de la inv
norma_inversa = np.linalg.norm(inversa, ord=np.inf)

condiconamiento = norma * norma_inversa

print(condiconamiento)
```
